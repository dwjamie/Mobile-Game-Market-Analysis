# 爬取手游信息：

包括名称、评分、总评论数、简介、一些评论、评论id、评论的用户名。

```python
#初步定义爬取top120的信息
N_top=120
```

```python
import xlwt
# 创建工作簿
f = xlwt.Workbook()

# 创建一个sheet
sheet1 = f.add_sheet('data', cell_overwrite_ok=True)

col_list=["游戏名称","评分","总评论数","简介","评论id","评论用户","评论内容"]

for j in range(len(col_list)):
    sheet1.write(0, j, col_list[j])
#初始化excel

row_index=1#记录下一行写入的位置
```

```python
'''
爬取数据并写入excel
'''
N_com_page=5#评论数据的页数

for i in range(0,N_top,30):
    top_url="https://www.taptap.com/ajax/top/download?total="+str(i)+"&page="+str(int(i/30+1))
    print(top_url)
    req1=requests.get(top_url)
    texts=json.loads(req1.text)#从json到字典
    texts=texts['data']
    texts=texts['html']#从字典到str

    re_top_href='''<div class="top-card-middle">\n\s*.*? href=(.*?) class="card-middle-title ">\n'''
    top_href=re.findall(re_top_href,texts,re.S)
    
    for j in top_href:
        url_game=j[1:-2]#游戏详情界面
        req2=requests.get(url_game)
        re_intro='''<div class="body-description-paragraph" id="description">\n\s*(.*?)\s*</div>'''
        intro=re.search(re_intro,req2.text)
        intro=intro.group(1)#游戏简介
        
        
        for k in range(N_com_page):
            url_comment=url_game+"/review?order=default&page="+str(int(k+1))+"#review-list"#评论页
            req3=requests.get(url_comment)
            
            
            #评论id和用户名
            re_help='<li id="review-(.*?)" class="taptap-review-item collapse in"\n    data-user="(.*?)">'#评论id和用户名
            find1=re.findall(re_help,req3.text)
            #注意要转化为list
            
            
            #评论总数
            re_help1='data-taptap-tab="review">评价\n                      <small>(.*?)</small>\n '
            find2=re.search(re_help1,req3.text)
            
            #评分
            re_help2='<span itemprop="ratingValue">(.*?)</span>\n'
            find3=re.search(re_help2,req3.text)
            
            #游戏名称
            re_help3='<title>(.*?) - 玩家评价 | TapTap 发现好游戏</title>\n'
            find4=re.search(re_help3,req3.text)
            
            #评论内容
            re_help4='''<div class="item-text-body" data-review-([0-9]*)="contents">\n\s*(.*?)\s*</div>'''
            find5=re.findall(re_help4,req3.text,re.S)   
            
            
            for line in range(len(find1)):#构造每一行的输入
                if row_index==65535:#如果超过上限则停止输入
                        break
                list_input=[find4.group(1),find3.group(1),find2.group(1),intro]
                list_input.extend(list(find1[line]))
                list_input.append(list(find5[line])[1])
                
                
                
                for colindex in range(len(col_list)):
                    sheet1.write(row_index, colindex, list_input[colindex])
                
                row_index=row_index+1
```

```python
f.save(save_path+"taptap.xls")
```

# 爬version2

```python
#初步定义爬取top150的信息
N_top=150
```

```python
def deencode(a):#调整编码 并忽略无法编码的词 方便后续处理
    return a.encode('gbk', 'ignore').decode('gbk')
```

```python
import xlwt
# 创建工作簿
f = xlwt.Workbook()

# 创建一个sheet
sheet1 = f.add_sheet('data', cell_overwrite_ok=True)

col_list=["游戏名称","总评分","总评论数","厂商","类别","标签","简介","评论id","评论用户","评论内容","评论星级"]

for j in range(len(col_list)):
    sheet1.write(0, j, col_list[j])
#初始化excel

row_index=1#记录下一行写入的位置
```

```python
'''
爬取数据并写入excel
'''
N_com_page=5#评论数据的页数

for i in range(0,N_top,30):
    top_url="https://www.taptap.com/ajax/top/download?total="+str(i)+"&page="+str(int(i/30+1))
    print(top_url)
    req1=requests.get(top_url)
    texts=json.loads(req1.text)#从json到字典
    texts=texts['data']
    texts=texts['html']#从字典到str
    
    #寻找url
    re_top_href='''<div class="top-card-middle">\n\s*.*? href=(.*?) class="card-middle-title ">\n'''
    top_href=re.findall(re_top_href,texts,re.S)
    #寻找厂商：
    re_indus='厂商:&nbsp;(.*?)</a>'
    find_indus=re.findall(re_indus,texts)
    #寻找类别：
    re_class='''<div class="card-middle-category">\n\s*<a href=.*? >(.*?)</a>\n'''
    find_class=re.findall(re_class,texts)
    #寻找标签：分两步走
    re_tag1='''<div class="card-tags">(.*?)</div>\n'''
    find_tag1=re.findall(re_tag1,texts,re.DOTALL)
    re_tag2='''>(.*?)</a>'''
    tag_list=[]
    for retag in find_tag1:
        find_tag2=re.findall(re_tag2,retag)
        tag_list.append(find_tag2)
    
    for j in range(len(top_href)):
        url_game=top_href[j][1:-2]#游戏详情界面
        req2=requests.get(url_game)
        re_intro='''<div class="body-description-paragraph" id="description">\n\s*(.*?)\s*</div>'''
        intro=re.search(re_intro,req2.text)
        intro=intro.group(1)#游戏简介
        
        industry=find_indus[j]#厂商
        game_class=find_class[j]#类别
        tag=tag_list[j]#标签
        
        for k in range(N_com_page):
            url_comment=url_game+"/review?order=default&page="+str(int(k+1))+"#review-list"#评论页
            req3=requests.get(url_comment)
            
            
            #评论id和用户名
            re_help='<li id="review-(.*?)" class="taptap-review-item collapse in"\n    data-user="(.*?)">'#评论id和用户名
            find1=re.findall(re_help,req3.text)
            #注意要转化为list
            
            #评论星级
            re_comstar='''<i class="colored" style="width: ([0-9]*?)px"></i>\n'''
            find_comstar=re.findall(re_comstar,req3.text)
            starlist=[]
            for px in find_comstar:
                starlist.append(int(px)/14)
            
            #评论总数
            re_help1='data-taptap-tab="review">评价\n                      <small>(.*?)</small>\n '
            find2=re.search(re_help1,req3.text)
            
            #评分
            re_help2='<span itemprop="ratingValue">(.*?)</span>\n'
            find3=re.search(re_help2,req3.text)
            
            #游戏名称
            re_help3='<title>(.*?) - 玩家评价 | TapTap 发现好游戏</title>\n'
            find4=re.search(re_help3,req3.text)
            
            #评论内容
            re_help4='''<div class="item-text-body" data-review-([0-9]*)="contents">\n\s*(.*?)\s*</div>'''
            find5=re.findall(re_help4,req3.text,re.S)   
            
            
            for line in range(len(find1)):#构造每一行的输入
                if row_index==65535:#如果超过上限则停止输入
                        break
                list_input=[find4.group(1),find3.group(1),find2.group(1),industry,game_class,tag,deencode(intro)]
                list_input.append(list(find1[line])[0])#评论id
                list_input.append(deencode(list(find1[line])[1]))#用户名
                list_input.append(deencode(list(find5[line])[1]))#评论内容
                list_input.append(starlist[line])
                
                
                
                for colindex in range(len(col_list)):
                    sheet1.write(row_index, colindex, list_input[colindex])
                
                row_index=row_index+1
```

```python
f.save(save_path+"taptap.xls")
```

```python
comment=taptap_data['评论内容']
```

```python
def presub(comment):#预处理函数
    com1=re.sub('[<p>]', '', comment)
    com2=re.sub('[</p>]', '', com1)
    com3=re.sub('[a-zA-Z]', '', com2)
    return com3
```

```python
subcom=comment.apply(presub)
```

```python
subcom.head()
```

```python
taptap_data['compre']=subcom
taptap_data.head()
```

```python
def presubintro(intro):
    intro1=re.sub('[<br>]', '', intro)
    intro2=re.sub('[</br>]', '', intro1)
    intro3=re.sub('[a-zA-Z]', '', intro2)
    return intro3
```

```python
taptap_data['intropre']=taptap_data['简介'].apply(presubintro)
taptap_data.head()
```
